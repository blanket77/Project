```python
# 글로벌 자산을 대표하는 ETF 데이터
import pandas as pd
import yfinance as yf
import pymysql
from sqlalchemy import create_engine
```

```python
################# tickers가 매개변수 
tickers = [
    'SPY', # 미국 주식
    'TLT', # 미국 장기채 
    'IEF', # 미국 중기채
    'GLD', # 금
    'DBC' # 상품

    # 'SPY',  # 미국 주식
    # 'IEV',  # 유럽 주식 
    # 'EWJ',  # 일본 주식
    # 'EEM',  # 이머징 주식
    # 'TLT',  # 미국 장기채
    # 'IEF',  # 미국 중기채
    # 'IYR',  # 미국 리츠
    # 'RWX',  # 글로벌 리츠
    # 'GLD',  # 금
    # 'DBC'  # 상품
]
#################

starts = '2022-07-01'
ends = '2022-12-30'
select_w = {} # 데이터프레임으로 받아야한다.

###################################### 언제부터 시작할거냐
all_data = {}
for ticker in tickers:
    all_data[ticker] = yf.download(ticker, start=starts, end=ends)
######################################
prices = pd.DataFrame(
    {tic: data['Adj Close']
     for tic, data in all_data.items()})

engine = create_engine('mysql+pymysql://root:1234@127.0.0.1:3306/stock_db')
prices.to_sql(name='sample_etf_close', con=engine, index=True, if_exists='replace')
engine.dispose()
```

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('mysql+pymysql://root:1234@127.0.0.1:3306/stock_db')
price = pd.read_sql('select * from sample_etf_close;', con=engine)
price = price.set_index(['Date'])
engine.dispose()

price.head()
```

```python
import bt

# 전체 자산 동일비중, 매월 말 리밸런싱
strategy = bt.Strategy("Asset_EW", [
       
    # bt.algos.SelectWhere(select_w),
    bt.algos.SelectAll(),
    bt.algos.WeighEqually(),
    bt.algos.RunOnce(),  ######### 몇주기, 몇달???
    bt.algos.Rebalance()
])
```

```python
data = price.dropna()

# 백테스트 생성
backtest = bt.Backtest(strategy, data)

# 백테스트 실행
result = bt.run(backtest)
```

```python
result.display()
```

```python
strategy = bt.Strategy("Asset_EW", [
       
    bt.algos.SelectWhere(select_w),
    bt.algos.WeighEqually(),
    bt.algos.RunDaily(),  ######### 몇주기, 몇달???
    bt.algos.Rebalance()
])

data = price.dropna()

# 백테스트 생성
backtest = bt.Backtest(strategy, data)

# 백테스트 실행
result = bt.run(backtest)

result.display()
```
