```python
# 글로벌 자산을 대표하는 ETF 데이터
import pandas as pd
import yfinance as yf
import pymysql
from sqlalchemy import create_engine

################# tickers가 매개변수 
tickers = [
    'SPY', # 미국 주식
    'TLT', # 미국 장기채 
    'IEF', # 미국 중기채
    'GLD', # 금
    'DBC' # 상품
    , 'MSFT', 'AAPL', 'NVDA', 'GOOG', 'GOOGL', 'AMZN', 'META', 'LLY', 'AVGO', 'TSLA', 'V', 'JPM', 'XOM', 'WMT', 'UNH', 'MA', 'PG', 'JNJ', 'HD', 'COST', 'MRK', 'ORCL', 'CVX', 'BAC', 'ABBV', 'KO', 'CRM', 'NFLX', 'AMD', 'PEP', 'ADBE', 'TMO', 'WFC', 'LIN', 'QCOM', 'DIS', 'ACN', 'MCD', 'CSCO', 'TMUS', 'ABT', 'GE', 'DHR', 'INTU', 'AMAT', 'CAT', 'AXP', 'TXN', 'VZ', 'AMGN', 'PFE', 'IBM', 'MS', 'PM', 'CMCSA', 'UBER', 'BX', 'UNP', 'NEE', 'NOW', 'COP', 'GS', 'NKE', 'SCHW', 'ISRG', 'RTX', 'SPGI', 'LOW', 'INTC', 'MU', 'ETN', 'HON', 'UPS', 'SYK', 'PGR', 'ELV', 'T', 'BKNG', 'LRCX', 'C', 'BLK', 'DE', 'LMT', 'TJX', 'MDT', 'BA', 'BSX', 'REGN', 'VRTX', 'ABNB', 'CB', 'ADI', 'ADP', 'MMC', 'PLD', 'CI', 'PANW', 'KLAC', 'MDLZ', 'FI'
    # 'SPY',  # 미국 주식
    # 'IEV',  # 유럽 주식 
    # 'EWJ',  # 일본 주식
    # 'EEM',  # 이머징 주식
    # 'TLT',  # 미국 장기채
    # 'IEF',  # 미국 중기채
    # 'IYR',  # 미국 리츠
    # 'RWX',  # 글로벌 리츠
    # 'GLD',  # 금
    # 'DBC'  # 상품
]
#################

starts = '2010-01-01'
ends = '2024-05-15'
select_w = {} # 데이터프레임으로 받아야한다.

###################################### 언제부터 시작할거냐
all_data = {}
for ticker in tickers:
    all_data[ticker] = yf.download(ticker, start=starts, end=ends)
######################################
prices = pd.DataFrame(
    {tic: data['Adj Close']
     for tic, data in all_data.items()})

engine = create_engine('mysql+pymysql://root:1234@127.0.0.1:3306/stock_db')
prices.to_sql(name='sample_etf_adj_close', con=engine, index=True, if_exists='replace')
engine.dispose()
```
